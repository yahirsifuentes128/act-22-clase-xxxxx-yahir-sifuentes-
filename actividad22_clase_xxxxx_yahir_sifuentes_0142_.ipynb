```python
class Joyero:
    def __init__(self, nombre, detalle, especialidad, nivel_paciencia):
        self.nombre = nombre
        self.atencion_al_detalle = detalle
        self.reparacion_y_restauracion = especialidad
        self.paciencia_y_concentracion = nivel_paciencia

    def acabado_y_pulido(self):
        print(f"\n✨ [{self.nombre}]: Iniciando pulido. Detalle: {self.atencion_al_detalle}.")

    def engastado_de_gemas(self):
        print(f"💎 [{self.nombre}]: Colocando gemas con paciencia {self.paciencia_y_concentracion}.")

    def tasacion_y_asesoramiento(self):
        print(f"🔍 [{self.nombre}]: Tasando pieza de {self.reparacion_y_restauracion}.")

# --- Inicio del Programa Interactivo ---
print("=== SISTEMA DE GESTIÓN DEL TALLER DE JOYERÍA ===")

# Lista para guardar a nuestros joyeros
taller = []

while True:
    print("\n--- Registro de Nuevo Joyero ---")

    # Captura de datos
    nom = input("Nombre del joyero: ")
    det = input(f"¿Cómo es la atención al detalle de {nom}?: ")
    esp = input(f"¿Cuál es su especialidad?: ")
    pac = input(f"Nivel de paciencia (Baja/Media/Alta/Extrema): ")

    # Creamos el objeto y lo guardamos en la lista
    nuevo_joyero = Joyero(nom, det, esp, pac)
    taller.append(nuevo_joyero)

    continuar = input("\n¿Deseas registrar a otro joyero? (si/no): ").lower()
    if continuar != "si":
        break

# --- Ejecución de la Jornada Laboral ---
print("\n=== INICIANDO JORNADA LABORAL EN EL TALLER ===")

for j in taller:
    print(f"\n>>> Turno de: {j.nombre}")
    j.acabado_y_pulido()
    j.engastado_de_gemas()
    j.tasacion_y_asesoramiento()

print("\n--- Fin del programa |  programa realizado por Yahir Sifuentes NL=56 ---")
```
